## Pridobivanje podatkov
Pridobljene podatke sem shranila v '.csv' datoteke za boljšo preglednost. To nam pomaga tudi pri nadalnjem obdelovanja, saj nam ni potrebno ponovno spletno pridobivanje.

Spodnja tabela služi za vse nadaljne, saj iz nje beremo naslednje podatke:
- temperatura,
- datum,
- mesto.

Pridobljeni podatki se posodabljajo za tekoči dan in segajo točno 365 dni v preteklost.

```python
import pandas as pd

df_vsa_mesta = pd.read_csv("podatki/zgodovina_vsa_mesta.csv")

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

df_vsa_mesta
```

## Povprečne dnevne temperature
V spodnjem grafu so prikazane povprečne dnevne temperature za vsako mesto, tekom leta.

Opazimo lahko, da ima Pariz najvišje povprečne temperature izmed treh mest, Berlin pa najnižje. Kljub temu so njune najnižje/najvišje temperature zelo podobne, kar je po vsej verjetnosti posledica njunih lokacij (nahajata se v sosednjih državah).

```python
from grafi import temperaturni_graf
mesta = ["Ljubljana", "Berlin", "Pārīz"]
temperaturni_graf(df_vsa_mesta, izbrana_mesta=mesta)
```

## Osnovna statistika
Tabela podaja minimalno, maksimalno in povprečno temperaturo ter standardni odmik za vsako mesto.

Ta statistična analiza ponovno pokaže podobnost med Parizom in Berlinom. Tudi Ljubljana ima zelo podobno statistiko, edino odstopanje se pokaže v zimskih mesecih, po vsej verjetnosti zaradi njene lege (kotlina).

```python
from grafi import statistika_mest
statistika_mest(df_vsa_mesta)
```

## Sezonska analiza
Prikazala sem tudi povprečne temperature glede na letni čas v tabeli in s stolpčnim diagramom.

Kljub sklepu iz prejšnjega predela, je sezonska analiza pokazala drugače. Čeprav ima Ljubljana najvišjo minimalno zabeleženo temperaturo, so v povprečju zime tam najhladnejše izmed treh mest. To pokaže, da v Parizu in Berlinu v zimskem času beležijo nižjo minimalno temperaturo, vendar je temperatura v povprečju še vedno višja.

Stolpični diagram te podatke prikaže še bolj pregledno, in so tako razliek v povprečnih temperaturah še bolj očitne.

```python
from grafi import sezonska_analiza
sezonska_analiza (df_vsa_mesta)
```

```python
from grafi import prikazi_stolpicni_diagram_letni_casi
sezonski_povprecji = grafi.sezonska_analiza(df_vsa_mesta)
sezonski_povprecji.loc[mesta]
prikazi_stolpicni_diagram_letni_casi(sezonski_povprecji, mesta)
```
